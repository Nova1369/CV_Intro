### Q1) Write a Python code that changes the colour value of the first 100X100 pixels of an image to black (0,0,0).

```python
img1=np.array(img)
for i in range(0,100,1):
    for j in range (0,100,1):
        img1[i,j,:]=0
        
plt.imshow(img1)
```

```python
def fun_black1(img1,k):
    flag=0
    row,col,cha=img1.shape
    for i in range(k,row,k):
        if flag%2==0:
            img1[i-k:i,:,0:3]=0
        flag+=1
    return img1
res=fun_black1(img,5)
plt.imshow(res)
```

```python
def fun_black2(img3,k):
    flag=0
    row,col,cha=img3.shape
    for i in range(k,col,k):
        if flag%2==0:
            img3[:,i-k:i,0:3]=0
        flag+=1
    return img3
res=fun_black2(img,5)
plt.imshow(res)
```

```python
def fun_black_row(img2,k):
    flag=0
    row,col,cha=img2.shape
    for i in range(k,col,k):
        if(flag%2==0):
            img2[:,i-k:i,0:3]=0
        flag+=1
    return img2

def fun_black_col(img2,k):
    flag=0
    row,col,cha=img2.shape
    for i in range(k,row,k):
        if(flag%2==0):
            img2[i-k:i,:,0:3]=0
        flag+=1
    return img2

def fun_black_both(img2,k):
    fun_black_row(img2,k)
    fun_black_col(img2,k)
    return img2

res=fun_black_both(img,5)
plt.imshow(res)
```

### Q3) Write an OpenCV code to load an image, extract a specific image region, and display both the original and extracted regions using Python.

```python
# Load an image from a file
img = cv2.imread('lena.jpg')

# Extract a specific region of interest (ROI) from the image
# For example, a 100x100 pixel square at the top-left corner
roi = img[0:100, 0:100]

# Display the original image in a window named 'Original'
cv2.imshow('Original', img)

# Display the extracted ROI in a window named 'ROI'
cv2.imshow('ROI', roi)

# Wait for a key press to exit
cv2.waitKey(0)

# Destroy all windows
cv2.destroyAllWindows()
```

### Q4) Implement a Python script that reads an image and calculates and plots the histogram of its grayscale version. Save the histogram plot as an image file

```python
# Read an image from a file
img = cv2.imread('lena.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Calculate the histogram of the grayscale image
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])

# Plot the histogram using matplotlib
plt.figure()
plt.title('Grayscale Histogram')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.plot(hist)
plt.xlim([0, 256])

# Save the histogram plot as an image file
plt.savefig('histogram.jpg')
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read an image from a file
img = cv2.imread('lena.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Calculate the histogram of the grayscale image using NumPy
hist, bins = np.histogram(gray.flatten(), bins=256, range=[0,256])

# Plot the histogram using matplotlib
plt.figure()
plt.title('Grayscale Histogram')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.plot(hist)
plt.xlim([0, 256])

# Save the histogram plot as an image file
plt.savefig('histogram.jpg')

# Show the plot
plt.show()
```

### Q5) Create a program that reads an image and applies a Gaussian blur to it using the filter shown in the image below. Display both the original image and the blurred image. (To apply a filter, you have to convolve the filter with the image)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
id_kernel = np.array([[1/16, 2/16, 1/16], [2/16, 4/16, 2/16], [1/16, 2/16, 1/16]])
flt_img = cv2.filter2D(src=img, ddepth=-1, kernel=id_kernel)
img8 = cv2.GaussianBlur(flt_img,(25,25),cv2.BORDER_DEFAULT)
plt.imshow(img8)
cv2.imwrite("gaussian.jpg", img)
```
